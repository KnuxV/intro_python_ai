Jusqu'à présent, nous nous sommes appuyés sur des fonctions prédéfinies dans le langage Python (comme les fonctions " print() " ou " input() ") ou que nous avions définies localement dans nos cellules de code (comme les fonctions " mon_affichage() " ou " ma_somme() " dans notre premier "notebook").

Au lieu de "réinventer la roue" chaque fois que l'on a besoin d'une fonctionnalité particulière qui n'est pas disponible par défaut dans le langage Python (et de perdre du temps et des ressources pour implémenter le code correspondant), on peut réutiliser le code que d'autres ont partagé avec la communauté, sous forme de "bibliothèques". 

C'est évidemment un gain de temps significatif et cela permet de simplifier énormément la phase de développement du code. <br>

## Comment faire appel aux bibliothèques ("librairies") ?

Mais, nous avons de la chance, il existe une bibliothèque qui pourrait nous aider :

datetime est une bibliothèque qui contient un ensemble d'objets et de fonctions utiles pour manipuler les dates...

Par exemple, la bibliothèque contient un objet "date" ui permet de définir et de manipuler les dates de manière standardisée.

## Rq:

Si nous savons que nous utiliserons fréquemment les fonctions d'une bibliothèque donnée, nous pouvons l'importer sous un alias plus court afin d'économiser de la frappe.

> <font color="red"><b>import</b></font> **bibliothèque** <font color="green"><b>as</b></font> <u>lib</u>

Pour appeler une fonction "func" (à titre d'exemple) à partir de la bibliothèque, il faut alors faire référence à la bibliothèque. Par exemple (en utilisant le raccourci précédent "lib" pour faire référence à la bibliothèque) :
> lib.func()

Une autre forme de l'instruction **import** permet d'importer des objets individuels de la bibliothèque directement dans la table de symboles de l'appelant :

> <font color="blue"><b>from</b></font> **bibliothèque** <font color="red"><b>import</b></font> object

### Voici un code qui utilise les fonctionnalités de datetime pour résoudre ce problème :`.

***

```python
# Nous commençons par importer l'outil dont nous avons besoin.
# Ici, nous allons utiliser un outil particulier (ici l'outil "date") de la bibliothèque "datetime" 
# (qui propose d'autres outils)

from datetime import date 

# Nous entrons la date (le format est année, mois, jour)
ma_date = date(2000, 1, 1)

# L'objet "date" a une fonction intégrée qui récupère la date actuelle :
aujourdhui = date.today() 

# les objets "date" peuvent être soustraits les uns des autres.
# pour calculer le temps écoulé entre deux dates, il suffit alors d'écrire :
nombre_de_jours = aujourdhui - ma_date

# Imprimons le résultat
print(nombre_de_jours)

# ou pour afficher seulement le nombre de jours: 
print("le nombre de jours :", nombre_de_jours.days)


# >> Pouvez-vous essayer de changer la variable "my_date" (essayez votre date d'anniversaire par exemple :-)
```

```python
# Fonction nb_jours_entre_deux_dates

from datetime import date 

def nb_jours_entre_deux_dates (date1, date2):
    
    nombre_de_jours = date2 - date1
    
    return nombre_de_jours.days
    

d1=date(1, 1, 1)
d2=date.today()
        
delta = nb_jours_entre_deux_dates (d1,d2)

print(delta)


# Peut-on afficher le nombre d'heures ?
# > récupérer le nb de jours dans une variable, multiplier par 24 et afficher le résultat

```

## Peut on visualiser le nb de jours sous forme d'une série de "points" ?

***

```python
# exemple : visualiser le nb de jours depuis sa naissance...

from datetime import date 

def print_dots (n):
    print("{} jours :".format(n))
    nb_points_par_ligne=50
    nb_lignes=n//nb_points_par_ligne
    for i in range(nb_lignes):
        print("o"*50)
    reste=n%50
    print("o"*reste)

aujourdhui=date.today()
date_naissance=date(1970,4,4)

nombre_de_jours = aujourdhui - date_naissance

print_dots(nombre_de_jours.days)


#print(nombre_de_jours.days)
```

## Pour aller plus loin (avec les palindromes...) :

#### Que faire avec une phrase du type :
> "L'âme sœur, elle, rue, ose mal." ?

Le problème de la "normalisation" des caractères:
- caractères non alphanumériques (blancs, ponctuation)
- caractères "spécuaix" (accents, etc.)

### Comment convertir cette chaine de caracatères pour détecter un palindrome ?
***


```python
texte="L'âme sœur, elle, rue, ose mal."

# "normaliser le texte"(remplacer les caractères accentués par leurs équivalents non accentués)

texte_alpha = ''.join(filter(str.isalnum, texte))
print(texte_alpha)

# "normaliser le texte"(remplacer les caractères accentués par leurs équivalents non accentués)

import unidecode
def normalise_texte(texte):
    texte=texte.lower()
    mot_sans_accents = unidecode.unidecode(texte)
    return mot_sans_accents

texte_normalise = normalise_texte(texte_alpha)
print (texte_normalise)

```
